Il est maintenant temps de mettre ces données sous formes de diagrammes. Nous allons commencer par faire des diagrammes circulaires.
Pour cela mettons nos données sous forme de listes qui pourront être prises en entrée des fonction de la librairie matplotlib.

```python
data_p1 = [empreinte_p1["logement"], empreinte_p1["voiture"], empreinte_p1["avion"], empreinte_p1["numérique"], empreinte_p1["alimentation"], empreinte_p1["textile"]+empreinte_p1["électroménager"]+empreinte_p1["autre produits"], empreinte_p1["services sociétaux"]]
labels_p1 = "Logement", "Voiture", "Avion", "Numérique", "Alimentation", "Divers", "Services sociétaux"

data_p2 = [empreinte_p2["logement"], empreinte_p2["train"], empreinte_p2["voiture"], empreinte_p2["numérique"], empreinte_p2["alimentation"], empreinte_p2["textile"]+empreinte_p2["électroménager"]+empreinte_p2["autre produits"], empreinte_p2["services sociétaux"]]
labels_p2 = "Logement", "Train", "Voiture", "Numérique", "Alimentation", "Divers", "Services sociétaux"
```

Maintenant, produisons les diagrammes circulaires grace à la librairie matplotlib.

```python
#Diagrammes circulaires.
fig1, ax1 = plt.subplots()

colors1 = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(data_p1)))
colors2 = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(data_p2)))

ax1.pie(data_p1, labels=labels_p1, colors=colors1, autopct='%1.1f%%', pctdistance=.7, 
       wedgeprops={"linewidth": 1, "edgecolor": "white"})
ax1.axis('equal')
plt.savefig('piedd')

fig2, ax2 = plt.subplots()
ax2.pie(data_p2, labels=labels_p2, colors=colors2, autopct='%1.1f%%', pctdistance=.8, 
       wedgeprops={"linewidth": 1, "edgecolor": "white"})#, frame=True)
ax2.axis('equal')
plt.savefig('pieam')

fig1.tight_layout()
fig2.tight_layout()
plt.show()
```

Ces diagrammes permettent de voir quelles sont les proportions de chaque catégorie dans notre empreinte d'équivalents $CO_2$. La discussion de ces graphique se trouve dans le travail remis.

Nous pouvons faire un diagramme en 3 dimensions pour comparer directement les deux empreintes $CO_2e$ directement selon les catégories.

```python
data1 = np.array([empreinte_p1["logement"], empreinte_p1["transport"], empreinte_p1["numérique"], empreinte_p1["alimentation"], empreinte_p1["textile"]+empreinte_p1["électroménager"]+empreinte_p1["autre produits"], empreinte_p1["services sociétaux"]])
data2 = np.array([empreinte_p2["logement"], empreinte_p2["transport"], empreinte_p2["numérique"], empreinte_p2["alimentation"], empreinte_p2["textile"]+empreinte_p2["électroménager"]+empreinte_p2["autre produits"], empreinte_p2["services sociétaux"]])
labels = "Logement", "Transport", "Numérique", "Alimentation", "Divers", "Services sociétaux"

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection = "3d")

countries = labels
data = np.array([data1,
                 data2])
years = [" ", " ", " ", "Damien", " ", " ", " ", "Anne-Marie"]

numOfCols = len(data1)
numOfRows = 2

xpos = np.arange(0, numOfCols, 1)
ypos = np.arange(0, numOfRows, 1)
xpos, ypos = np.meshgrid(xpos + 0.5, ypos + 0.5)

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(numOfCols * numOfRows)

dx = np.ones(numOfRows * numOfCols) * 0.5
dy = np.ones(numOfCols * numOfRows) * 0.5
dz = data.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz)
ax.set_xticklabels(countries)
ax.set_yticklabels(years)

#ax.set_xlabel('Countries')
#ax.set_ylabel('Over the years')
ax.set_zlabel('kgCO2e')

plt.show()
```

Bien que l'effet 3D soit à mon goût, cette vue n'est pas pratique pour une comparaison simple des émissions d'équivalents carbone des deux personnes.

Essayons alors de faire un diagramme en batonnets.

```python
index = np.arange(len(data1))
bar_width = 0.4

fig, ax = plt.subplots(figsize=(14,6)) 
ax.bar(index, data1, bar_width,
                label="Damien", color=plt.get_cmap('Blues')(0.4))

ax.bar(index+bar_width, data2,
                 bar_width, label="Anne-Marie")

ax.set_xlabel('Categories')
ax.set_ylabel('kgCO2e/an')
ax.set_title("Empreinte d'équivalent CO2")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

Cette vue est bien meilleure pour une comparaison directe. De plus nous pouvons aisément passer l'ordonnée à l'échelle logarithmique pour voir les différences d'ordres de grandeurs des différentes catégories.

```python
fig, ax = plt.subplots(figsize=(14,6)) 
ax.bar(index, data1, bar_width,
                label="Damien", color=plt.get_cmap('Blues')(0.4))

ax.bar(index+bar_width, data2,
                 bar_width, label="Anne-Marie")

ax.set_xlabel('Categories')
ax.set_ylabel('kgCO2e/an')
ax.set_title("Empreinte d'équivalent CO2")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()
ax.set_yscale('log')

plt.show()
```

Cette échelle permet de constater que la catégorie "numérique" qui comprend l'utilisation non professionnelle d'internet et les appareils électroniques à un impact d'équivalents $CO_2$ un ordre de grandeur plus petit que celui des autres catégories.

Procédons maintenant à quelques ajustements pour rendre ce diagramme plus clair.

```python
#Remaniement des données pour stacker les différentes contributions à une catégorie.
data1dd = np.array([empreinte_p1["logement"], empreinte_p1["voiture"], empreinte_p1["numérique"], empreinte_p1["repas"], empreinte_p1["textile"]+empreinte_p1["électroménager"]+empreinte_p1["autre produits"], empreinte_p1["services sociétaux"]])
data2dd = np.array([0,                        empreinte_p1["avion"],   0, empreinte_p1["boisson"]+empreinte_p1["déchets"],  0                                                                        ,          0])
data1am = np.array([empreinte_p2["logement"], empreinte_p2["voiture"], empreinte_p2["numérique"], empreinte_p2["repas"], empreinte_p2["textile"]+empreinte_p2["électroménager"]+empreinte_p2["autre produits"], empreinte_p2["services sociétaux"]])
data2am = np.array([0,                       empreinte_p2["train"],     0,              empreinte_p2["boisson"]+empreinte_p2["déchets"],  0                                                                     , 0])
labels1 = " ", "Voiture", " ", "Repas", " ", " "
labels2dd = " ", "Avion", " ", "Reste", " ", " "
labels2am = " ", "Train", " ", "Reste", " ", " "

#Diagramme en batonnets.
index = np.arange(len(data1dd))
bar_width = 0.4

fig, ax = plt.subplots(figsize=(14.5,8))

ax1 = ax.bar(index, data1dd, bar_width,
                label="Damien", color=plt.get_cmap('Blues')(0.4), ec="white")
ax2 = ax.bar(index, data2dd, bar_width,
                color=plt.get_cmap('Blues')(0.4), ec='white', bottom=data1dd)

ax3 = ax.bar(index+bar_width, data1am,
                 bar_width, label="Anne-Marie", ec="white", color=plt.get_cmap('Blues')(0.8))
ax4 = ax.bar(index+bar_width, data2am,
                 bar_width, ec="white", color=plt.get_cmap('Blues')(0.8))

ax.bar_label(ax1, labels=labels1, label_type='center', fontweight="bold")
ax.bar_label(ax3, labels=labels1, label_type='center', fontweight="bold")
ax.bar_label(ax2, labels=labels2dd, label_type='center', fontweight="bold")
ax.bar_label(ax4, labels=labels2am, label_type='center', fontweight="bold")


ax.set_xlabel('Categories')
ax.set_ylabel('kgCO2e/an')
ax.set_title("Empreinte d'équivalent CO2")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()
#ax.set_yscale('log')


plt.savefig("Empreinte d'équivalent CO2")
plt.show()
```

Nous pouvons normaliser ce graphique pour faire apparaitre les proportions que prennent chaque catégorie dans nos émissions de gaz à effet de serre.

```python
#Remaniement des données pour stacker les différentes contributions à une catégorie.
data1dd = data1dd/data1.sum()
data2dd = data2dd/data1.sum()
data1am = data1am/data2.sum()
data2am = data2am/data2.sum()
labels1 = " ", "Voiture", " ", "Repas", " ", " "
labels2dd = " ", "Avion", " ", "Reste", " ", " "
labels2am = " ", "Train", " ", "Reste", " ", " "

#Diagramme en batonnets.
index = np.arange(len(data1dd))
bar_width = 0.4

fig, ax = plt.subplots(figsize=(14.5,8))

ax1 = ax.bar(index, data1dd, bar_width,
                label="Damien", color=plt.get_cmap('Blues')(0.4), ec="white")
ax2 = ax.bar(index, data2dd, bar_width,
                color=plt.get_cmap('Blues')(0.4), ec='white', bottom=data1dd)

ax3 = ax.bar(index+bar_width, data1am,
                 bar_width, label="Anne-Marie", ec="white", color=plt.get_cmap('Blues')(0.8))
ax4 = ax.bar(index+bar_width, data2am,
                 bar_width, ec="white", color=plt.get_cmap('Blues')(0.8))

ax.bar_label(ax1, labels=labels1, label_type='center', fontweight="bold")
ax.bar_label(ax3, labels=labels1, label_type='center', fontweight="bold")
ax.bar_label(ax2, labels=labels2dd, label_type='center', fontweight="bold")
ax.bar_label(ax4, labels=labels2am, label_type='center', fontweight="bold")


ax.set_xlabel('Categories')
ax.set_ylabel('Proportion des émission de CO2e (%)')
ax.set_title("Empreinte d'équivalent CO2 normalisée")
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()
#ax.set_yscale('log')


plt.savefig("Empreinte d'équivalent CO2 normalisée")
plt.show()
```
